```julia
initial_data_type = :chichon_2D

#Grid parameters
N_out = 50
M_out = 50
const N_Fields_out = 14
Lx_out = 100.0
Ly_out = 100.0
lengthpars_out = (Lx_out, Ly_out, N_out, M_out, N_Fields_out, mink_meric!)
```

```julia
#initial data vector
initial_data = ones(N_out,M_out,2*N_Fields_out);


C0_out = 10
C1_out = 10
C2_out = 10

χ_out = (1.0, 1.0)
C_out = (C0_out, C1_out, C2_out)

tol_out = 1e-15
iter_max_out = 10

par_inidat, par_source, par_flux, par_inv, par_ev = parameters_creator(χ_out, C_out, lengthpars_out, tol_out, iter_max_out, mink_metric!)

```
